```python
population,profit = data[:,0],data[:,1]
population=population[:,np.newaxis]
profit = profit[:,np.newaxis]
population.shape
m=len(population)
ones=np.ones((m,1))
population=np.hstack((ones,population))
print(population.shape)
print(population[:5,:])

#population.shape
#t = np.ones((1,len(population)))
#t.shape
#population=np.concatenate((t,population),axis=0)
```

```python
theta = np.zeros((2,1))
print(theta)                 
iterations = 1500
alpha = 0.01
def cost_function(x, y, theta):
    m = len(y)
    cost = np.sum(np.square(x.dot(theta) - y))
    return (1 / (2*m)) * cost
```

```python
def gradientDescent(X, y, theta, alpha, iterations):
    cost=[]
    for _ in range(iterations):
        
        temp = np.dot(X, theta) - y
        cost.append(np.sum(temp))
        temp = np.dot(X.T, temp)
        theta = theta - (alpha/m) * temp
    return theta,cost
theta,cost = gradientDescent(population,profit, theta, alpha, iterations)
print(theta)
```

```python
plt.scatter(population[:,1], profit, label='Traning Data')
plt.plot(population[:,1], population.dot(theta), 'r-', label='Linear Regression')
plt.legend()
plt.xlabel('Population')
plt.ylabel('Profit')
```

```python
plt.plot(np.arange(iterations), cost,label="No. if Iterations",c='r')
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.legend()
```
